# creating multiline graph plot

```python
x1 = [11, 22, 33]
x2 = [47, 85, 69]
y1 = [11, 22, 33]
y2 = [75, 23, 16]
```

```python
plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
from matplotlib import style
```

```python
# style.use('ggplot')
style.use('dark_background')
```

```python
plt.plot(x1, y1, label='sarso', linewidth=4) #linewidth increases the thickness of the line
plt.plot(x2, y2, label='taad')
plt.legend() #to show the labels in the chart
```

```python
plt.scatter(x1, y1)
plt.scatter(x2, y2)
```
